###### Using Mean, Median And Mode To Estimate Values.
<hr>

* This might not be the actual values but it is weighed to see which one  gives more accuracy.

```python
Mean_FA = Data["fixed acidity"].mean()
Mean_VA = Data["volatile acidity"].mean()
Mean_CA = Data["citric acid"].mean()
Mean_C = Data["chlorides"].mean()
Mean_FSD = Data["free sulfur dioxide"].mean()
Mean_TSD = Data["total sulfur dioxide"].mean()
Mean_D = Data["density"].mean()
Mean_PH = Data["pH"].mean()
Mean_Q = Data["quality"].mean()
```

```python
print(f'Mean_FA = {Mean_FA}')
print(f'Mean_VA = {Mean_VA}')
print(f'Mean_CA = {Mean_CA}')
print(f'Mean_C = {Mean_C}')
print(f'Mean_FSD = {Mean_FSD}')
print(f'Mean_TSD = {Mean_TSD}')
print(f'Mean_D = {Mean_D}')
print(f'Mean_PH = {Mean_PH}')
print(f'Mean_Q = {Mean_Q}')

```

##### Strategising The Data Given With The Mean

* filling the emoty set above with the **_MEAN_**

```python
Data['fixed acidity'] = Data['fixed acidity'].fillna(Mean_FA)
Data['volatile acidity'] = Data['volatile acidity'].fillna(Mean_VA)
Data['citric acid'] = Data['citric acid'].fillna(Mean_CA)
Data['chlorides'] = Data['chlorides'].fillna(Mean_C)
Data['free sulfur dioxide'] = Data['free sulfur dioxide'].fillna(Mean_FSD)
Data['total sulfur dioxide'] = Data['total sulfur dioxide'].fillna(Mean_TSD)
Data['density'] = Data['density'].fillna(Mean_D)
Data['pH'] = Data['pH'].fillna(Mean_PH)
Data['quality'] = Data['quality'].fillna(Mean_Q)
```

##### As a statistical analyst, a detialed look at the  data would imply a look at it from the mean, mode and median.
* A critical look at the data would imply performing a mathematical grouping on the production of the data with the mean, mode anf meadian values to select the set of values that gives an upgrade in the production of red wine 🍷.

```python
groups = Data[['quality','chlorides','sulphates',]].groupby('quality')
groups

fig,ax = plt.subplots(figsize=(15,10))

for quality, group in groups:
    ax.scatter(group['chlorides'], group['sulphates'], marker = 'x', label=quality )
    
plt.show()
    
```

* For comparison of the dataset above the **chlorides** and **sulphates** from both mean version and median set of data were selected respectively to compare the set of values with more efficiency using **THE COmPARISON OF DATASET WITH THE SCATTER DIAGRAM APPROACH**. 

```python
groups = Data[['quality','chlorides','chlorides',]].groupby('quality')
groups

fig,ax = plt.subplots(figsize=(15,10))

for quality, group in groups:
    ax.scatter(group['chlorides'], group['chlorides'], marker = 'x', label=quality )
    
plt.show()
```

* for a more vivid look, **_chlorides_** of both the **mean** and **median** version of dataset were use to have more graphical sense of the data to choose which is best fit for _increase in production_.
